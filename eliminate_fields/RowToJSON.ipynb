```python
import pandas as pd
import json

import math


def removenans(entry):
    if isinstance(entry, str):
        if entry =='NaN':
            return None
        else:
            return entry
    elif math.isnan(entry):
        return None
    else:
        return entry

#takes in a row as argument
def row_to_JSON(row):
    ein=row["EIN"]
    name=row["NAME"]
    url=row["URL"]
    phone=row["Phone"]
    services=row["NTEE_CD"] 
    hours=row["HOURS"] 
    hoursDumped = json.dumps(hours)
    
    phone = removenans(phone)
    name = removenans(name)
    if isinstance(hours, float):
        hours = None
    else:
        hours = removenans(hours)
    url = removenans(url)
    services = removenans(services)
    
    
    #return dict
    return {'name': name, 'url':url, 'phone': phone, 'services':services, 'hours':hours}
```
